```python
import numpy as np
```

```python
# 1. Знайти матрицю С= 𝐴𝐵−𝐵𝐴:

A = np.array([[1, 2],[4, -1]])

B = np.array([[2, -3],[-4, 1]])

AB = np.dot(A, B)
BA = np.dot(B, A)

C = AB - BA

print("Матриця A:\n", A)
print("\nМатриця B:\n", B)
print("\nМатриця AB:\n", AB)
print("\nМатриця BA:\n", BA)
print("\nРезультат обчислення C = AB - BA:")
print(C)
```

8. Розв’язати систему лінійних рівнянь методом Крамера, матричним
методом (тобто обчислити обернену матрицю) і м. Гауса. Потім
зробити перевірку за допомогою solve() пакету linalg :

```python
a = np.array([
    [3,-5, 3],
    [0, 2, 0],
    [2, 7, 0],
])
b = np.array([[1], [4], [8]])
```

```python
def solve_cramer(a, b, verbose=False):
  det = np.linalg.det(a)
  if verbose: print(f"Визначник матриці А: {round(det)}")
  if det == 0:
    print("Визначення дорівнює нулю - розв'язків немає")
    return None

  len_a = len(a)
  x = np.zeros((len_a, 1))
  for i in range(len_a):
    temp = np.copy(a)
    temp[:, [i]] = b
    x[i, 0] = np.linalg.det(temp)/det
  return x

print(f"Вектор рішення :\n{solve_cramer(a, b)}")
```

```python
N = 4
M = 4

A = np.random.randint(0, 11, size=(N, M))

print("Згенерована матриця A:")
print(A)
```

```python
# Обчислення середнього для кожного рядка
row_means = np.mean(A, axis=1)

print("Середні значення по рядках:")
for i, mean in enumerate(row_means):
    print(f"Рядок {i+1}: {mean:.2f}")
```

```python
# Обчислення середнього для кожного стовпця
col_means = np.mean(A, axis=0)

print("Середні значення по стовпцях:")
for j, mean in enumerate(col_means):
    print(f"Стовпець {j+1}: {mean:.2f}")
```
